```python
import sys
sys.path.append(r'C:\Users\Said\Desktop\proyecto-ds-plantilla\src')

from anonymize import DataAnonymizer
import pandas as pd

# ============================================================
# S3.2 — Usando DataAnonymizer para Leonali
# ============================================================

# Cargar dataset
df_leonali = pd.read_parquet(r'C:\Users\Said\Documents\Datasets\Alimentos\Dataset1.parquet')

print("ANTES:")
print(f"Shape: {df_leonali.shape}")
print(f"Columnas: {df_leonali.columns.tolist()}")
print(f"Primeras 2 IDs: {df_leonali['IdCliente'].head(2).tolist()}")

# Crear instancia del anonymizer
anon = DataAnonymizer(salt="leonali_2024_secret")

# Anonimizar
df_clean = anon.anonymize_dataset(
    df_leonali,
    drop_cols=['NombreCliente2'],
    hash_cols={'IdCliente': 12},
    datetime_cols={'FechaProceso': '%d/%m/%Y'}
)

print("\nDESPUÉS:")
print(f"Shape: {df_clean.shape}")
print(f"Primeras 2 IDs (hasheados): {df_clean['IdCliente'].head(2).tolist()}")
print(f"\nDataset limpio y anonimizado")

# Guardar dataset anonimizado
df_clean.to_parquet(r'C:\Users\Said\Documents\Datasets\Alimentos\Dataset1_anonimizado.parquet', index=False)

print(f"✓ Dataset guardado: Dataset1_anonimizado.parquet")
print(f"✓ Tamaño: {df_clean.memory_usage(deep=True).sum() / 1e6:.2f} MB")
print(f"✓ Listo para análisis")
```
