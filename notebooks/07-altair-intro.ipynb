```python
# ===========================================================
# F1M1 — Módulo 1: EDA Leonali
# Notebook 07 — Introducción a Altair: gramática de gráficos
# ===========================================================

import pandas as pd
import altair as alt

df = pd.read_parquet('../data/processed/Dataset1_anonimizado.parquet')
df.shape
```

```python
datos = df.groupby('Categoria1')['Costo_Venta'].sum().reset_index()

chart0=alt.Chart(datos).mark_bar().encode(
    x='Categoria1',
    y='Costo_Venta'
)
chart0.save('../reports/07_barras_altair.png')
```

```python
ventas_año = df.groupby(df['FechaProceso'].dt.year)['Costo_Venta'].sum().reset_index()

chart1=alt.Chart(ventas_año).mark_line(point=True).encode(
    x='FechaProceso:O',
    y='Costo_Venta:Q'
)
chart1.save('../reports/07_lineas_altair.png')
```

```python
muestra = df.sample(3000, random_state=42)
```

```python
muestra = df.sample(3000, random_state=42)

chart2=alt.Chart(muestra).mark_circle().encode(
    x='Piezas_Entregadas',
    y='Costo_Venta',
    color='Categoria1'
)
chart2.save('../reports/07_circle_altair.png')
```

```python
muestra = df.sample(3000, random_state=42)

chart3=alt.Chart(muestra).mark_bar().encode(x=alt.X('Costo_Venta',bin=True),y='count()')
chart3.save('../reports/07_barras_altair2.png')
```

```python
ventas_cat_año = df.groupby([df['FechaProceso'].dt.year, 'Categoria1'])['Costo_Venta'].sum().reset_index()

chart4=alt.Chart(ventas_cat_año).mark_rect().encode(
    x='FechaProceso:O',
    y='Categoria1:N',
    color='Costo_Venta:Q'
)
chart4.save('../reports/07_heatmap_altair.png')
```

```python
top10 = df.groupby('Producto')['Costo_Venta'].sum().reset_index().sort_values('Costo_Venta', ascending=False).head(10)

chart5=alt.Chart(top10).mark_bar().encode(
    x='Costo_Venta:Q',
    y=alt.Y('Producto:N', sort='-x'))
chart5.save('../reports/07_barras_hor_altair.png')
```
