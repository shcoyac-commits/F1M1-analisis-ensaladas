```python
# ===========================================================
# F1M1 — Módulo 1: EDA Leonali
# Notebook 09 — Análisis ABC y Pareto
# ===========================================================

import pandas as pd
import plotly.express as px
import sys
sys.path.insert(0, '..')
from src.kpis import pareto_80_20

df = pd.read_parquet('../data/processed/Dataset1_anonimizado.parquet')
df.shape
```

```python
clientes = df.groupby('IdCliente').agg(
    gasto_total = ('Costo_Venta', 'sum'),
    frecuencia = ('FechaProceso', 'nunique'),
    variedad = ('IdProducto', 'nunique')
)
```

```python
clientes.head(10
              )
```

```python
clientes.describe()
```

```python
clientes['seg_gasto'] = pd.qcut(clientes['gasto_total'], q=3, labels=['Bajo', 'Medio', 'Alto'])
clientes['seg_frecuencia'] = pd.qcut(clientes['frecuencia'], q=3, labels=['Bajo', 'Medio', 'Alto'])
```

```python
clientes['seg_gasto'].value_counts()
```

```python
pd.crosstab(clientes['seg_gasto'], clientes['seg_frecuencia'])
```

```python
clientes['segmento'] = clientes['seg_gasto'].astype(str) + ' - ' + clientes['seg_frecuencia'].astype(str)
clientes['segmento'].value_counts()
```

```python
canal_cliente = df.groupby('IdCliente')['Canal2'].agg(lambda x: x.mode()[0]).reset_index()
canal_cliente.columns = ['IdCliente', 'Canal2']
```

```python
clientes2 = clientes.reset_index().merge(canal_cliente, on='IdCliente')
```

```python
# Alto - Alto
print('=== ALTO - ALTO ===')
print(clientes2[clientes2['segmento'] == 'Alto - Alto']['Canal2'].value_counts())

# Bajo - Alto
print('\n=== BAJO - ALTO ===')
print(clientes2[clientes2['segmento'] == 'Bajo - Alto']['Canal2'].value_counts())
```

```python
clientes = df.groupby('IdCliente').agg(
    gasto_total = ('Costo_Venta', 'sum'),
    frecuencia = ('FechaProceso', 'nunique'),
    variedad = ('IdProducto', 'nunique')
)
```

```python
fig = px.bar(
    clientes['segmento'].value_counts().reset_index(),
    x='segmento',
    y='count',
    color='segmento',
    title='Distribución de Clientes por Segmento - Leonali'
)
fig.show()
fig.write_image('../reports/10_segmentos_clientes.png', width=1200, height=500)
```
