```python
# Lista de productos de una farmacia
productos = ["paracetamol", "ibuprofeno", "vitamina C", "agua", "café"]
print(productos)
print(type(productos))
```

```python
# Acceder a elementos por posición
print(productos[0])   # el primero
print(productos[1])   # el segundo
print(productos[-1])  # el último (truco útil)
```

```python
# Agregar un producto
productos.append("refresco")
print(productos)

# Quitar un producto
productos.remove("café")
print(productos)

# Cuántos productos hay
print(len(productos))
```

```python
# Un ticket de farmacia
ticket = {
    "id_cliente": 1042,
    "producto": "paracetamol",
    "cantidad": 2,
    "precio_unitario": 25.50,
    "categoria": "medicamento"
}

print(ticket)
print(ticket["producto"])      # acceder por llave
print(ticket["precio_unitario"])
```

```python
# Calcular el total y agregarlo al ticket
ticket["total"] = ticket["cantidad"] * ticket["precio_unitario"]
print(ticket["total"])
print(ticket)
```

```python
# Lista de tickets del día
tickets_del_dia = [
    {"id_ticket": 1001, "articulo": "paracetamol", "cantidad": 2, "precio_unitario": 25.50, "categoria": "medicamento"},
    {"id_ticket": 1002, "articulo": "café americano", "cantidad": 1, "precio_unitario": 18.0, "categoria": "café"},
    {"id_ticket": 1003, "articulo": "ibuprofeno", "cantidad": 1, "precio_unitario": 35.0, "categoria": "medicamento"},
    {"id_ticket": 1004, "articulo": "agua", "cantidad": 3, "precio_unitario": 15.0, "categoria": "bebida"},
]

# Cuántos tickets hubo
print("Total tickets:", len(tickets_del_dia))

# Ver el segundo ticket
print("Ticket 2:", tickets_del_dia[1])

# Categorías únicas del día
categorias_dia = set(ticket["categoria"] for ticket in tickets_del_dia)
print("Categorías vendidas:", categorias_dia)
```
